```python
import sys
import os
import pickle
import pandas as pd
import numpy as np
from functools import reduce  

%config InlineBackend.figure_format = 'svg'
```

```python
stat2 = pickle.load(open("../data/interim/stats2.pickle","rb"))
data2 = pickle.load(open("../data/processed/data2.pickle","rb"))
```

```python
data2c = data2.copy()
data2c.head()
```

```python
stat2=stat2.rename(columns={'MP_x':'MP'});
```

```python
stat2['Player'] = stat2['Player'].map(lambda x: x.replace('.',""))
```

```python
totallist=['MP', 'FG', 'FGA','3P', '3PA',  '2P', 
           '2PA', 'FT', 'FTA', 'ORB', 'DRB', 
           'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']
```

```python
for i in range(data2c.shape[0]):
    curr = data2c.iloc[i].Player
    curryear = data2c.iloc[i].Year
    years  = [curryear, curryear-1, curryear-2]
    threeyrs = stat2[(stat2.Player == curr) & (stat2.Year.isin(years))]
    if threeyrs.shape[0] > 1:
        print("Update row "+str(i))
        for stat in totallist:
            data2c.iloc[i, data2c.columns.get_loc(stat)] = (reduce((lambda x, y: x + y), 
                              [k*v for k,v in zip(threeyrs[stat],threeyrs.G)])) #/ threeyrs.G.sum()
        data2c.iloc[i, data2c.columns.get_loc('G')] = threeyrs.G.sum()#/len(threeyrs)
        data2c.iloc[i, data2c.columns.get_loc('GS')] = threeyrs.GS.sum()#/len(threeyrs)
    
```
