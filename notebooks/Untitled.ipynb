```python
import pandas as pd

df = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'], 
                   'A': ['A2', 'A1', 'A2', 'A3', 'A4', 'A5'], 'B':['A4', 'A2', 'B7', 'C3', 'D4', 'B5']})
```

```python
df
```

```python
other = pd.DataFrame({'key': ['K0', 'K1', 'K2','K4'],'A': ['B0', 'B1', 'A2','A4']})
```

```python
other
```

```python
COLS = ['key','A']

df[~df[COLS].isin(other[COLS].to_dict(
    orient='list')).all(axis=1)]
```

```python
df.merge(other, indicator=True, how='outer')
```

```python
merged = df.merge(other, indicator=True, how='outer')
merged[merged['_merge'] == 'left_only']
```
